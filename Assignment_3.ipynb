```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",data=iris)
```

```python
plt.scatter(iris["sepal_width"],iris["petal_width"])
```

```python
sns.lineplot(x="sepal_length",y="petal_width",data=iris)
```

```python
sns.distplot(iris["petal_length"])
```

```python
sns.barplot(y="sepal_width",x="species",data=iris)
```

```python
sns.boxplot(x="sepal_length",y="petal_length",data=iris)
```

```python
sns.catplot(x="species",y="sepal_width",data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot=True)
```
